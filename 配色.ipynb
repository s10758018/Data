```
NBA1314<-fetch_NBAPlayerStatistics("13-14")
#載入NBA2013~2014的資料
```

```
NBA13140rder<-NBA1314[order(NBA1314$FieldGoalsMade,decreasing=T),]
NBA13140rder[1:30,]#依FieldGoalsMade的由大到小排出30名
```

```
subset(NBA1314,Team=="TOR")#搜尋TOR隊的欄位資料
```

```
NBA1314[grepl("Robin",NBA1314$Name),]#字串收尋名字裡有Robin的球員
```

```
qplot(TotalRebounds,TotalMinutesPlayed,data=NBA1314)
#縱軸是TotalRebounds，橫軸是TotalMinutesPlayed的散佈圖
```

```
qplot(TotalRebounds,
 TotalMinutesPlayed,
  data = NBA1314,
  color=Position)
  #縱軸是TotalRebounds，橫軸是TotalMinutesPlayed的散佈圖，
  #依照Position分類塗色
```

```
 qplot(TotalRebounds, data = NBA1314, fill = Position)
 #用TotalRebounds畫出直方圖，依照Position分類塗色
```

```
qplot(TotalRebounds,data=NBA1314,
      facets=Position~., binwidth=50)
      #用TotalRebounds畫的直方圖，一圖分開呈現多筆資料，長條寬度每格代表50
```
