```
pokemon <- read.csv("寶可夢能力值表.csv")
```

```
pokemon
```

```
pokemon %>% count(屬性)
#顯示每種屬性各有幾種寶可夢
```

```
pokemon %>% count(屬性, sort = T)
#由最多種到最少種，列出各屬性有幾種寶可夢
```

```
pokemon %>% 
  count(屬性) %>% 
  ggplot(aes(x = 屬性, y = n, fill = 屬性))+
  geom_col() +
  geom_text(aes(label = n), vjust = -0.25) + 
  labs(title = "各屬性有幾種寶可夢", x = "屬性", y = "該屬性的寶可夢數量") + 
  theme(legend.position = "none")
  #用柱狀圖畫出各屬性有幾種寶可夢，可看到水系103種最多，幽靈系26種最少）
```

```
pokemon %>% 
  count(屬性) %>% 
  ggplot(aes(x = fct_reorder(屬性, n), y = n, fill = 屬性))+
  geom_col() +
  geom_text(aes(label = n), vjust = -0.25) + 
  labs(title = "各屬性有幾種寶可夢", x = "屬性", y = "該屬性的寶可夢數量") + 
  theme(legend.position = "none")
    #用柱狀圖，由少到多依序排出每屬性各有幾種寶可夢，幽靈系最少種，水系最多種
```

```
pokemon %>%
count(攻擊) %>%
ggplot(aes(x = 攻擊, y = n))+
geom_col()
#柱狀圖攻擊力分析，攻擊力能力值在100~250之間的寶可夢種類最多，攻擊力達到300的有2種
```

```
pokemon %>%
group_by(屬性) %>%
summarise(mean_攻擊 = mean(攻擊, na.rm = T))
#分析各屬性的平均攻擊力，龍系平均攻擊力最高，蟲系平均攻擊力最低
```

```
pokemon %>%
mutate(屬性 = fct_reorder(屬性, 攻擊, median, na.rm = TRUE)) %>%
ggplot(aes(屬性, 攻擊))+
geom_point()+
geom_boxplot()+
coord_flip()
#箱形圖分析各屬性寶可夢的攻擊力分佈，龍系攻擊力集中且偏高，
#岩石系的最高和最低攻擊力差異最大，
#只有超能力系的攻擊力分佈觸及300的標線，因此可以知道攻擊力分析柱狀圖裡達到300的2種寶可夢都是屬於超能力系
```
