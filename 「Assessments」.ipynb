```
library(SportsAnalytics)
library(ggplot2)
```

```
NBA1314<-fetch_NBAPlayerStatistics("13-14")
```

```
str(NBA1314)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1314)
qplot(GamesPlayed,FieldGoalsMade,data=NBA1314)
```

```
qplot(FieldGoalsAttempted, TotalPoints, data = NBA1314,color=Position)
qplot(GamesPlayed, FieldGoalsMade, data = NBA1314,color=Position)
```

```
 qplot(TotalPoints, data = NBA1314, fill = Position)
 qplot(FieldGoalsMade, data = NBA1314, fill = Position)
```

```
qplot(FieldGoalsAttempted,TotalPoints,
      data=NBA1314,
      facets=Position~.)
qplot(GamesPlayed,FieldGoalsMade,
      data=NBA1314,
      facets=Position~.)
```
